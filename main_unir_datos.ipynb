```python
# Unir los datos
data.columns  = ["tspan_num","Salinidad"]
```

```python
data["tspan"] = mdates.num2date(data["tspan_num"])
data["tspan"] = data["tspan"].dt.tz_localize(None)
```

```python
data.head()
```

```python
# Ahora que data (los datos corregidos) y data_BMT3_3 (los datos originales) tienen la misma columna de tiempo, podemos unirlos.
# 1. Colocar a tspan como índice en data (los datos corregidos)
data.set_index("tspan", inplace=True)
```

```python
# 2. Hacer un dataframe temporal a partir de los datos corregidos que tenga solo las fechas y los datos NaN.
data_solo_nans = data[data["Salinidad"].isna()]
data_solo_nans.head()
```

```python
#3. Ahora debo hacer una copia del df de datos reales
datos_reales = data_BMT3_3.copy()
datos_reales.loc[data_solo_nans.index, ["Salinidad","Conductividad"]] = np.nan
datos_reales[["Salinidad","Conductividad"]].head(20) # Verifico que se hayan puesto NaN en las fechas correctas
```

```python
# Guardar el DataFrame corregido en un archivo .pkl usando pickle
# Especifica el nombre del archivo
nombre_archivo = 'msj24_validados_realt_2025120100_2025123123.pkl'

# Guardar el objeto en el archivo (usando 'with' para asegurar que el archivo se cierre automáticamente)
with open(nombre_archivo, 'wb') as archivo:
    pickle.dump(datos_reales, archivo)
print(f"¡Datos guardados exitosamente en '{nombre_archivo}'!")
```
